```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
digits.target
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df=pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['target']=digits.target
```

```python
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3, random_state=10)
```

Create KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

I tried various values of K manually and K=5 seem to be giving me the best score

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,5))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted",fontsize=20)
plt.ylabel('Truth',fontsize=20)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
